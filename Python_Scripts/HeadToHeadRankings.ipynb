# Evaluate head-to-head pairings

Convert to numpy

```python
data = data.to_numpy()
```

Comparison Algorithm

```python
def headToHead(cand1, cand2):
  c1Total = 0
  c2Total = 0

  #loop through all rows
  for i in range(data.shape[0]):
    #loop through candidates in rows
    for j in range(1, data.shape[1]):
      #if a candidate is listed before another candidate
      #that candidate is prefered
      #all votes in row belong to preferred candidate
      if data[i][j] == cand1:
        c1Total = c1Total + data[i][0]
        break;
      elif data[i][j] == cand2:
        c2Total = c2Total + data[i][0]
        break;

  #return candidate pairing and votes accumulated
  return [cand1 + " vs. " + cand2, c1Total, c2Total]
```

Make matrix to hold comparison data

```python
rankings = []
```

Loop through candidate combinations

```python
for i in range(candidates.shape[0]):
  for j in range(i + 1, candidates.shape[0]):
    rankings.append(headToHead(candidates[i], candidates[j]))
```

Check Rankings

```python
for ranking in rankings:
  print(ranking)
```
