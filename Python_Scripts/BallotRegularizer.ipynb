# Download Ballot dataframe from CSV file

```python
filename = 'Portland_06142022_Schoolboarddistrict5'

data = pd.read_csv('/content/drive/MyDrive/RCV Voting Method Research/Voting Ballots/' + filename + '.csv')
data.tail()
```

Change Column Headings after index (Inclusive)

```python
index = 1

for i in range(index, len(data.columns)):
  name = data.columns[i]
  data.rename(columns={name:"rank" + str(i - index + 1)}, inplace = True)
```

Remove Rows

```python
start = 2123
end = 2131

data = data.drop(data.index[start:end])
```
