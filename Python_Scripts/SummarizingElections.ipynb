Get names of all files

```python
files = [f for f in os.listdir('.') if os.path.isfile(f)]
```

Define output

```python
output = pd.DataFrame(columns = ["Place", "Start", "End", "Count"])
```

Summarize elections

```python
for f in files:
  place = ""
  year = "20"
  findingPlace = True

  for char in f:
    if char.isdigit():
      findingPlace = False
    elif not findingPlace:
      break

    if findingPlace:
      place += char
    else:
      year += char

  year = int(year)

  if output["Place"].str.contains(place).sum() > 0:
    output.loc[output["Place"] == place, "Count"] = output.loc[output["Place"] == place, "Count"] + 1

    if year < output.loc[output["Place"] == place, "Start"].sum():
      output.loc[output["Place"] == place, "Start"] = year
    elif year > output.loc[output["Place"] == place, "End"].sum():
      output.loc[output["Place"] == place, "End"] = year

  else:
    row = [place, year, year, 1]
    output.loc[len(output)] = row

output.to_csv("../Summary.csv")
```
